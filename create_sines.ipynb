This file edits the outputs previously saved from Bajes to fix the discontinuities and undifferentiable points in the amplitude and phase. This also edits the files to make sure the phase all starts at the same value for every file - hoping this makes the GAN's job easier.

```python
import os
import csv
import numpy as np
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
# mount google drive for reading and writing files

from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
# make directory to save sims. this will later be zipped

Path('/sims').mkdir(exist_ok=True)
```

```python
# create simulations; save relevant data about amplitude and frequency to metadata file

res = 1024   # how many datapoints to generate

sim = 0

times = [i for i in range(res)]

simsarray = []
freqarray = []
ampsarray = []

for freq in range(1,101):
    omega = np.pi*2*freq/(res)
    for j in range(0,101):
        sim += 1
        amp = (1 + 0.01*j)/3
        wave = amp*np.sin(omega*np.arange(0, res, 1))

        df = pd.DataFrame({'time': times, 'wave': wave})
        simsarray.append('/sims/' + str(sim) + '.csv')
        freqarray.append(freq)
        ampsarray.append(amp)

        df.to_csv('/sims/' + str(sim) + '.csv', sep= '\t', index = False)
        
mdat = pd.DataFrame({'sim': simsarray, 'freq': freqarray, 'amp': ampsarray})
mdat.to_csv('/content/drive/MyDrive/Chapman/510/Project/METADATA.csv', sep= '\t', index = False)
```
