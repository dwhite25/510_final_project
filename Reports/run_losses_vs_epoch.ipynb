```python
import csv
import numpy as np
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
outpath = 'drive/MyDrive/Chapman/510/Project/Outs/512n_0.01_0.01_b'

sb.set_theme(style='whitegrid')

filepath = outpath + '/weights/losses.csv'
losses = pd.read_csv(filepath, header=None)
losses = losses.to_numpy(dtype=np.float32)

fig, (sp1) = plt.subplots(1, figsize=(14,7))
sp1.plot(losses[:,0], losses[:,1], label='Generator')
sp1.plot(losses[:,0], losses[:,2], label='Discriminator')
# sp1.set_ylim(0, 50)
sp1.legend()
plt.show()
```
